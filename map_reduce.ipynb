### Config 1 - 1 master 1 slave

```python
# 1 master 1 slave - Gutenberg
from datetime import datetime
start_time = datetime.now()

!hadoop jar /usr/lib/hadoop/hadoop-streaming.jar -files  mapper.py,reducer.py \
    -input "/data/Gutenberg/*" \
    -output /out/Gutenberg_1_1 \
    -mapper ./mapper.py \
    -reducer ./reducer.py

end_time = datetime.now()
print('Job execution time',str(end_time - start_time))
```

### Config 4 - 1 master 9 slave 30 reducers

```python
# 1 master 9 slave 9 reducers - wikipedia
from datetime import datetime
start_time = datetime.now()

!hadoop jar /usr/lib/hadoop/hadoop-streaming.jar -files  mapper.py,reducer.py \
    -D mapreduce.job.reduces=30 \
    -input "/data/wikipedia/*" \
    -output /out/wikipedia_1_9_30 \
    -mapper ./mapper.py \
    -reducer ./reducer.py

end_time = datetime.now()
print('Job execution time',str(end_time - start_time))
```
